First, let's look at the python program we previously used to find all the prime numbers between 1 and 100.


````python
for x in range (1,100):
    status = True
    for y in range (2,x):
        if (x % y) == 0:
            status = False
    if (status == True):
        print (x)
````

Here's the updated version.  Notice the `count` variable.

````python
count = 0
for x in range (1,100):
    status = True
    for y in range (2,x):
        if (x % y) == 0:
            status = False
    if (status == True):
        count = count + 1
print ("there are",count,"prime numbers")
````
